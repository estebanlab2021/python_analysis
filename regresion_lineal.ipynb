## **REGRESIÓN LINEAL**

Al igual que el ANOVA y las Correlaciones, la Regresión lineal se basa en una relación LINEAL, pero entre una variable dependiente y una o varias variables independientes.

Busca la relación entre dos variables numéricas, pero tratando de predecir una en función de la otra, es decir, predecir o estimar la variable dependiente a partir de la variable independiente.

El concepto original de Regresión fue introducido por primera vez por Sir Francis Galton a finales del siglo 19, haciendo estudios poblacionales sobre la talla

## GENERALIDADES

Actualmente, el significado de regresión en Estadística, equivale en general al término función en matemáticas, es decir, dado el valor de una variable X, le corresponde un valor de la variable Y.

Esta relación o función lineal, se calcula a partir de un procedimiento llamado rectas de mejor ajuste, con una técnica llamada mímimos cuadrados, originalmente propuesto por Carl Friedrich Gauss. Es decir, la idea es generar una “recta” que mejor se aproxime a los datos de acuerdo con el criterio de mínimo error cuadrático.

## ERROR MÍNIMO CUADRÁTICO

El cuadrado de un error es el cuadrado de un residuo, por lo tanto, el cuadrado de un error es igual a la diferencia entre el valor real y el valor ajustado por el modelo de regresión elevado a la dos.

Por eso el criterio de los mínimos cuadrados también se llama criterio del mínimo error cuadrático.

Los modelos de Regresión nos permiten establecer estimaciones o predicciones de determinados valores, con cierta confiabilidad o certeza.

En el análisis de regresión se realizan dos operaciones principales:

1.  Obtener una Ecuación que represente la relación entre las variables (ecuación de regresión) y obtener la línea de regresión dada por esta ecuación. (pueden ser lineales o no)

2.  Estimar una variable llamada variable Dependiente o Respuesta, a partir de otra variable o grupo de variables llamadas variables Independientes o Explicativas.



## DIFERENTES MODELOS DE REGRESIÓN

Regresión lineal simple
$\bar{Y}=a+bX$

Regresión lineal múltiple
$\bar{Y}=a+b_1X_1+b_2X_2+...+b_nX_n$

Regresión no lineal:
*  Exponencial: $\bar{Y}=a*e^{bX}$
*  Potencia:  $\bar{Y}=a*X^{b}$
*  Cuadrático:  $\bar{Y}=a+b_1X+b_2X^2$
*  etc

## SUPUESTOS DE LA REGRESIÓN

El modelo de regresión asume ciertos supuestos o condiciones de tipo matemático, que deben ser tomados en consideración antes de asignarle un valor predictivo. De esta forma, se puede verificar si los resultados sean válidos y robustos.

En la práctica, rara vez se cumplen, o se puede demostrar que se cumplen todas, sin embargo esto no significa que el modelo no sea útil. Lo importante es ser consciente de ellas y del impacto que esto tiene en las conclusiones que se extraen del modelo.

Estos supuestos tienen que ver con la distribución de los residuales.

Distinguimos varios supuestos principales:

1.  Normalidad de los residuales
2.  No colinealidad o multicolinealidad
3.  Homocedasticidad de los residuales
4.  Independencia de los residuales (o no autocorrelación)
5.  Tamaño de la muestra

## NORMALIDAD DE LOS RESIDUALES

Para facilitar la estimación del modelo de regresión es exigible la normalidad de la distribución de los errores

Se pueden aplicar pruebas de normalidad a la distribución de los residuales,  como Kolmogórov-Smirnov o Shapiro-Wilk. Sin embargo, una alternativa bastante utilizada para este caso es la prueba de Jarque-Bera.

La Prueba de Jarque-Bera parte del hecho de que una distribución normal tiene coeficiente de asimetría igual a 0 y Curtosis igual a O. Estos dos elementos se miden a partir de los residuos del modelo lineal.

Tambien suelen usarse los gráfico Q-Q Plots o histogramas

En caso de rechazar la normalidad, esta se puede corregir, realizando transformaciones fundamentalmente a la variable dependiente.

Una de las transformaciones más usadas es la Transformación de Box-Cox. Esta es una transformación potencial que corrige la asimetría de una variable, varianzas diferentes o la no linealidad entre variables. En consecuencia, resulta muy útil para transformar una variable y obtener una nueva que siga una distribución normal.


## NO COLINEALIDAD O MULTICOLINEALIDAD

En los modelos lineales múltiples, los predictores deben ser independientes, no debe de haber colinealidad entre ellos. La colinealidad ocurre cuando un predictor está linealmente relacionado con uno o varios de los otros predictores del modelo. En consecuencia, si hay colinealidad o multicolinealidad, no se puede identificar de forma precisa el efecto individual que tiene cada predictor sobre la variable respuesta.

No existe un método estadístico concreto para determinar la existencia de colinealidad o multicolinealidad entre los predictores de un modelo de regresión, sin embargo, se han desarrollado numerosas reglas prácticas que tratan de determinar en qué medida afectan al modelo. Los pasos recomendados a seguir son:

Si el coeficiente de determinación  R2 es alto pero ninguno de los predictores resulta significativo, hay indicios de colinealidad.
Verificar que en la matriz de correlación de los predictores, no existan correlaciones muy altas.

Otra forma de evaluar la colinealidad es a través del Factor de Inflación de la Varianza (VIF), esto lo que quiere decir es que  si hay presencia de colinealidad la varianza se ve aumentada por ese efecto. Los límites de referencia para la interpretación son:

## HOMOCEDASTICIDAD DE LOS RESIDUALES

Debe existir homocedasticidad u homogeneidad de las varianzas

El supuesto de homocedasticidad exige que la varianza de la variable respuesta debe ser constante en todo el rango de los predictores. Para comprobarlo suelen representarse los residuos del modelo frente a cada predictor.

Si la varianza es constante, se distribuyen de forma aleatoria manteniendo una misma dispersión y sin ningún patrón específico.

También se puede recurrir a contrastes de homocedasticidad como el test de Breusch-Pagan.

## Cómo corregir la heterocedasticidad en caso que sea detectada:
*  Transformar la variable dependiente: Una forma de corregir la heterocedasticidad es transformar la variable dependiente, por ejemplo usando el log.

*  Redefinir la variable dependiente: Otra forma de corregir la heterocedasticidad es redefinir la variable dependiente. Una forma habitual de hacerlo es utilizar una tasa para la variable dependiente, en lugar del valor bruto.

*  Utilizar regresión ponderada: Otra forma de corregir la heterocedasticidad es utilizar la regresión ponderada. Este tipo de regresión asigna un peso a cada punto de datos en función de la varianza de su valor ajustado


## INDEPENDENCIA DE LOS RESIDUALES

No debe existir autocorrelación entre los residuales.

Los valores de cada observación son independientes de los otros.

Een este modelo, se supone que los residuos son independientes unos de otros (no están correlacionados).

Si se viola el supuesto de independencia, algunos resultados de ajustes de modelos pudieran no ser fiables. Por ejemplo, una correlación positiva entre los términos de error tiende a inflar los valores t de los coeficientes, haciendo que los predictores parezcan significativos cuando pudieran no serlo.

Para estos casos se usa el test de autocorrelation Durbin-Watson



## TAMAÑO DE LA MUESTRA

Otro elemento que puede afectar la validez del modelo es el tamaño de la muestra.

Sin embargo, no existe una condición establecida para el número mínimo de observaciones, pero para prevenir que una variable resulte muy influyente cuando realmente no lo es, se recomienda que la cantidad de observaciones sea entre 10 y 20 veces el número de variables independientes

# **ELEMENTOS DE INTERPRETACIÓN Y RESULTADOS**

## COEFICIENTE DE DETERMINACIÓN (R2)

De los residuos se calcula el Coeficiente de Determinación R2:

Este coeficiente determina Cuánto es la proporción de la varianza de la v. dependiente explicada por las v. independientes.

El R2 va de 0 (el modelo no explica la varianza de la v. dependiente) a 1 (el modelo explica toda la varianza de la v. dependiente).

Por su parte, el R2 ajustado es una medida corregida de bondad de ajuste (precisión de modelo) para los modelos lineales. Su uso es para evitar la sobreestimación del modelo, cuando se van añadiendo variables (penaliza el uso excesivo de variables).

## SIGNIFICACIÓN DEL MODELO

El modelo de regresión también presenta un valor F y un p-value, que representa la significación del modelo completo. Es decir, Si el p.valor es menor a 0.05 hay evidencias que el modelo es capaz de explicar a la variable dependiente.

Impacto individual de variables independientes sobre variable dependiente (p-valor) y coeficiente de regresión parcial(coeficiente b)

## METRICAS DE RENDIMIENTO DEL MODELO

*  **Error Absoluto Medio (MAE)**: Es la diferencia entre el valor pronosticado y el valor real en cada punto pronosticado

*  **Error Cuadrático Medio (MSE)**: Es la diferencia la cuadrado entre el valor real y pronosticado. Representa la distancia al cuadrado entre los valores reales y predichos.

*  **Raíz del Error Cuadrático Medio (RMSE)**: Es la raíz cuadrada de MSE. Mide la magnitud promedio de los errores y se ocupa de las desviaciones del valor real.

```python
# Librerías a utilizar
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import f_regression
from sklearn import metrics
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

## Cargamos los datos

```python
# Cargamos el fichero y lo convertimos en u dataframe
data = pd.read_csv('https://raw.githubusercontent.com/estebanlab2021/python_analysis/master/datos/ejemplo_python.csv', sep=";")
data.head()     #Verificamos las primeras filas de nuestra data con la función head()
```

Nos vamos a quedar con las variables Age e Income

```python
data_set = data.loc[:, ['Age','Income']]
print(data_set)
```

Declaramos los valores que estan fuera de rango como valores perdidos para poder identificarlos y eliminamos los valores perdidos.

```python
data_set.loc[data_set['Income'] < 0, ['Income']] = np.nan
data_set.loc[data_set['Income'] > 100, ['Income']] = np.nan
data_set.loc[data_set['Age'] > 100, ['Age']] = np.nan
```

```python
data_set = data_set.dropna(axis=0, how='any')
```

```python
data_set.shape  # verificamos la forma del nuevo dataframe
```

## Vemos la Correlación

```python
correlacion = stats.spearmanr(data_set)
print("La Correlación de Spearman: ", correlacion[0])
print("P-value: ", correlacion[1])
```

## Revisamos el gráfico de dispersión

```python
plt.scatter(x=data_set['Age'], y=data_set['Income'])
plt.show()
```

## Separación de los datos

Definimos la variable dependiente como Y y las variables independientes como X

```python
X = data_set['Age']      #Independiente
y = data_set['Income']   #Dependiente
```

```python
# Realizamos la división de los datos en train y test (entrenamiento y prueba)
X_train, X_test, y_train, y_test = train_test_split(X.values.reshape(-1,1), y.values.reshape(-1,1), train_size=0.7, random_state=0)
```

```python
# Verificamos la forma en que quedan nuestros datos de entrenamiento y test
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

## Generamos el modelo


```python
reg = LinearRegression().fit(X=X_train.reshape(-1,1), y = y_train.reshape(-1,1))
```

```python
# Información del modelo
print("Intercept:", reg.intercept_.round(3))
print("Coeficiente:", list(zip(pd.DataFrame(X).columns, reg.coef_.flatten()) ))
print("Coeficiente de determinación R^2:", reg.score(X.values.reshape(-1,1),y.values.reshape(-1,1)))
```

## Es decir, nuestro modelo tiene la siguiente ecuación:

*  Income = Intercept + (Coeficente *  Age)

*  Income = 12.66383 + (0.8982 * Age)

*  Por ejemplo, si Establecemos una edad de 20 años, entonces el "Income" estimado sería:

*  Income = 12.66383 + (0.8982 * 20) = 30.6295

En Python lo resolvemos de la siguiente forma:

```python
income = reg.predict(np.array([[20]]))
print(income)
```

## Valor F y p-value del modelo

```python
f_statistic, p_values = f_regression(X_train, y_train)
print(f_statistic)
print(p_values)
```

## Ahora vamos a representar la linea de regresión en el gráfico de dispersión

```python
x = X_train
y = y_train
#a = reg.intercept_
#b = reg.coef_.flatten()
plt.scatter(x, y)  #hacemos nuestro tradicional gráfico de dispersión
#b, a = np.polyfit(X_train.flatten(), y_train.flatten(), deg=1)   #Otra forma de obtener los coefientes del modelo
#plt.plot(x, b*x + a, color="red")  #Otra forma de añadir la linea al gráfico usando la ecuación del modelo
plt.plot(x, reg.predict(X = X_train), color="red") #Añadimos la linea al gráfico usando el predictor del modelo
plt.xlabel("Age")
plt.ylabel("Income")
plt.title("Gráfico de Dispersión con Recta de estimación")
plt.show()
```

## Con esto hemos entrenado el modelo. Ahora vamos a evaluar su capacidad predictiva empleando el conjunto de test.



```python
# Error de test del modelo

predicciones = reg.predict(X = X_test)

mae = metrics.mean_absolute_error(y_true= y_test, y_pred = predicciones)
print("El error mae es: ", mae)

mse = metrics.mean_squared_error(y_true= y_test, y_pred = predicciones)
print("El error mse es: ", mse)

rmse = metrics.mean_squared_error(y_true  = y_test, y_pred  = predicciones, squared = False)
print("El error rmse es: ", rmse)

r2 = metrics.r2_score(y_true=y_test, y_pred = predicciones)
print("El R^2 del modelo es: ", r2)

# R2 Ajustado
n = len(y_test) # Cantidad de Filas del test
k = X_test.shape[1] #Cantidad de Columnas (variables independientes)
r2_adj = 1-((1-r2)*(n-1)/(n-k-1))
print("El R^2 ajustado del modelo es: ", r2_adj)
```

## USANDO LA LIBRERÍA STATSMODELS

La implementación de regresión lineal de Statsmodels, es más completa que la de Scikitlearn ya que, además de ajustar el modelo, permite calcular los test estadísticos y análisis necesarios para verificar que se cumplen las condiciones sobre las que se basa este tipo de modelos.

Definimos el modelo de regresión lineal con los datos de entrenamiento previamente definidos

```python
#modelo = sm.OLS(y_train, sm.add_constant(X_train))
#linreg = modelo.fit()
linreg = sm.OLS(y_train, sm.add_constant(X_train)).fit()
print(linreg.summary())
```

## Verificamos los supuestos del modelo

```python
#Obtenemos los residuales
residuales = linreg.resid
#Normalidad de los residuos Jarque - Bera test

test = sm.stats.stattools.jarque_bera(residuales)
#statsmodels.stats.stattools.jarque_bera(residuales)
print("Test Jarque-Bera:", test[0])
print("p-value:", test[1])
```

```python
# Verificación gráfica de la normalidad de los residuos
fig = sm.qqplot(linreg.resid, line='45', fit=True)
plt.show()
```

```python
#Homocedasticidad de los residuos
test = sm.stats.diagnostic.het_breuschpagan(resid=residuales, exog_het=linreg.model.exog)
print("Test Breusch-Pagan: ",test[0])
print("p-value: ",test[1])
```

```python
#Independencia de los residuales
#Autocorrelación de Durbin-Watson test
# un valor de 2 indica que no hay autocorrelación
# un valor cercano a 0 indica autocorrelación positiva
# un valor cercano a 4 indica una autocorrelación negativa
# En general valores en un rango de 1,5 y 2,5 se consideran normales
test = sm.stats.stattools.durbin_watson(residuales)
test
```

```python
#No colinealidad de los predictores
variables = linreg.model.exog
vif = [variance_inflation_factor(variables, i)for i in range(variables.shape[1])]

vifs = pd.DataFrame({'variables':linreg.model.exog_names, 'vif':['%.2f' % elem for elem in vif]})
print(vifs)
```

## Obtenemos las Métricas

```python
#Obtenemos la predicción
y_pred_reg = linreg.predict(sm.add_constant(X_test))
```

```python
MAE = metrics.mean_absolute_error(y_true= y_test, y_pred = y_pred_reg)
print("El error mae es: ", MAE)

MSE = metrics.mean_squared_error(y_true= y_test, y_pred = y_pred_reg)
print("El error mse es: ", MSE)

RMSE = metrics.mean_squared_error(y_test, y_pred_reg, squared=False) #si squared es true MSE si es False RMSE
print('El error RMSE: %.2f' %RMSE)
```

# **REGRESION LINEAL MÚLTIPLE**

Definimos un nuevo conjunto de variables numéricas a partir de la base data

```python
data_set2 = data.loc[:, ['Age','Employ','Income','Creddebt', 'Debtinc','OthDebt']]
print(data_set2)
```

```python
#Nuevamente realizamos una "limpieza de los datos"
data_set2.loc[data_set2['Income'] < 0, ['Income']] = np.nan
data_set2.loc[data_set2['Income'] > 100, ['Income']] = np.nan
data_set2.loc[data_set2['Age'] > 100, ['Age']] = np.nan

data_set2 = data_set2.dropna(axis=0, how='any')

data_set2.shape  # verificamos la forma del nuevo dataframe
```

```python
#Separamos en la variable dependiente Y = Income y el resto como independientes X
y = data_set2['Income']
del data_set2['Income']
X = data_set2
```

Separamos el total de datos en datos de entrenamiento y de prueba

```python
# Realizamos la división de los datos en train y test (entrenamiento y prueba)
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=0)
```

Volvemos a generar el modelo

```python
modelo2 = sm.OLS(y_train, sm.add_constant(X_train))
linreg2 = modelo2.fit()
print(linreg2.summary())
```

Verificamos los supuestos

```python
#Normalidad de los residuos Jarque - Bera test
test = sm.stats.stattools.jarque_bera(linreg2.resid)
print("Test Jarque-Bera:", test[0])
print("p-value:", test[1])

#Homocedasticidad de los residuos
test = sm.stats.diagnostic.het_breuschpagan(resid=linreg2.resid, exog_het=linreg2.model.exog)
print("Test Breusch-Pagan: ",test[0])
print("p-value: ",test[1])

#Independencia de los residuales
#Autocorrelación de Durbin-Watson test

test = sm.stats.stattools.durbin_watson(linreg2.resid)
print("Test Durbin-Watson: ", test)
```

```python
# Verificación gráfica de la normalidad de los residuos
fig = sm.qqplot(linreg2.resid, line='45', fit=True)
plt.show()
```

```python
#No colinealidad de los predictores

variables = linreg2.model.exog
vif = [variance_inflation_factor(variables, i)for i in range(variables.shape[1])]

vifs = pd.DataFrame({'variables':linreg2.model.exog_names, 'vif':['%.2f' % elem for elem in vif]})
print(vifs)
```

Calculamos las métricas

```python
#Obtenemos la predicción
y_pred_reg = linreg2.predict(sm.add_constant(X_test))

MAE = metrics.mean_absolute_error(y_true= y_test, y_pred = y_pred_reg)
print("El error mae es: ", MAE)

MSE = metrics.mean_squared_error(y_true= y_test, y_pred = y_pred_reg)
print("El error mse es: ", MSE)

RMSE = metrics.mean_squared_error(y_test, y_pred_reg, squared=False)
print('El error RMSE: %.2f' %RMSE)
```
